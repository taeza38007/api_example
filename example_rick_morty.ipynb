```python
import requests, json 
import pandas as pd
```

```python
url = 'https://rickandmortyapi.com/api/'
endpoint = 'character'
```

```python
r = requests.get(url + endpoint)
print(r)
```

```python
data = r.json()
type(data)
```

```python
df = pd.DataFrame(data['results'])
df
```

```python
df.columns
df.to_json('rick.json')
```

```python
with open ('morty.json', 'w') as outfile:
    json.dump(data, outfile)
```
